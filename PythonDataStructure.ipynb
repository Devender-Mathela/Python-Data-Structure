Q2. Explain the difference between mutable and immutable data types with examples.
Ans:

Mutable: Mutable data types Can be changed after creation. Example: list = [1, 2, 3] -> list[0] = 10 is allowed.

Immutable:Immutable data types Cannot be changed after creation. Example: tuple = (1, 2, 3)->tuple[0] = 10(error).

Q19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Ans:

Dictionaries (before Python 3.7) didn't guarantee order because they use hashing. In modern Python (3.7+), they preserve insertion order, but retrieval is still by key, not by position.
